```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)
```

```python
plt.plot(x, y)

plt.xlabel("Milliseconds")
plt.ylabel("Awesomeness")

plt.show()
```
