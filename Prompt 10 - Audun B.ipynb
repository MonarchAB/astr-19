```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#I chose a Gaussian distribution as Gauss is the name of my friends band named after the mathemetician 
normalDistribution = np.random.normal(0, 1, 1000)
```

```python
plt.hist(normalDistribution, bins=100, color='lightgreen', edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Gaussian Histogram")

plt.savefig("NormalDistributionHistogramPlot.pdf")

plt.show()
```
